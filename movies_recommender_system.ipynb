```python
movies.dropna(inplace=True)
```

```python
import ast

def extract(a):
    l = []
    for i in ast.literal_eval(a):
        l.append(i['name'])
    return l
```

```python
movies['keywords']=movies['keywords'].apply(extract)
```

```python
movies['genres']=movies['genres'].apply(extract)
```

```python
import ast

def extract2(a):
    l = []
    counter =0
    for i in ast.literal_eval(a):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(extract2)
```

```python
import ast

def extract3(a):
    l = []
    for i in ast.literal_eval(a):
        if i['job']==['Director']:
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(extract2)
```
